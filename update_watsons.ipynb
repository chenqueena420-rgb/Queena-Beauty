# 確保所有美妝資料都包含屈臣氏通路

本筆記本會讀取 `products-data.json`，將 `屈臣氏` 加入每筆產品的 `available_at` 與 `links`，但不刪除任何現有資料。

```python
import json
import urllib.parse
from pathlib import Path

DATA_PATH = Path(r'c:\Users\queena\Desktop\my beauty\products-data.json')
```

```python
# Load Beauty Product Data
with DATA_PATH.open('r', encoding='utf-8') as f:
    products = json.load(f)

print(f'Loaded {len(products)} products')
print('Example product keys:', list(products[0].keys()))
```

```python
# Inspect Data Structure
empty_avail = [p['name'] for p in products if '屈臣氏' not in p.get('available_at', [])]
empty_links = [p['name'] for p in products if not p.get('links') or '屈臣氏' not in p.get('links', {}) or p['links'].get('屈臣氏', '') == '']
print('Products missing Watsons in available_at:', len(empty_avail))
print('Products missing Watsons link:', len(empty_links))
print('Sample missing available_at:', empty_avail[:10])
print('Sample missing links:', empty_links[:10])
```

```python
# Add Watsons Sales Channel
for product in products:
    if 'available_at' not in product or not isinstance(product['available_at'], list):
        product['available_at'] = []
    if '屈臣氏' not in product['available_at']:
        product['available_at'].append('屈臣氏')

    if 'links' not in product or not isinstance(product['links'], dict):
        product['links'] = {}
    if '屈臣氏' not in product['links'] or product['links'].get('屈臣氏','') == '':
        name = product.get('name','').strip()
        if name:
            search_url = 'https://www.watsons.com.tw/search?text=' + urllib.parse.quote(name)
            product['links']['屈臣氏'] = search_url

print('Watsons channel added to all products')
```

```python
# Verify Updated Data
missing_avail = [p['name'] for p in products if '屈臣氏' not in p.get('available_at', [])]
missing_links = [p['name'] for p in products if not p.get('links') or '屈臣氏' not in p.get('links', {}) or p['links'].get('屈臣氏','') == '']
print('After update - missing Watsons in available_at:', len(empty_avail))
print('After update - missing Watsons in links:', len(empty_links))
print('First 3 Watsons links:')
for p in products[:3]:
    print(p['name'], p['links'].get('屈臣氏'))
```

```python
# Save Updated Dataset
with DATA_PATH.open('w', encoding='utf-8') as f:
    json.dump(products, f, ensure_ascii=False, indent=2)
print('Saved updated products-data.json')
```
